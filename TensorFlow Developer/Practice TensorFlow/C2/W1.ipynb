```python
import tensorflow as tf

# Sample inputs
sentences = [
    'i love my dog',
    'I, love my cat loves'
    ]


# Initialize the layer
vectorize_layer = tf.keras.layers.TextVectorization(max_tokens=3)
```

```python
vectorize_layer.adapt(sentences)
print(vectorize_layer.get_vocabulary())


#for index, sentence in enumerate(vectorize_layer.get_vocabulary()):
#    print(f'{index}: {sentence}')
```
